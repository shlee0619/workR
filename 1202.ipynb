```python
# 독립표본 t검정
data <- data.frame(group1 = c(7,5,5,3,4,7,3,6,1,2,10,9,3,10,2,8,5,5,8,1,2,5,1,12,8,4,15,5,3,4),
                   group2 = c(5,3,4,4,2,3,4,5,2,5,4,7,5,4,6,7,6,2,8,7,8,8,7,9,9,5,7,8,6,6))
data
head(data)
str(data)
# 평균
g1_m <- mean(data$group1)
g2_m <- mean(data$group2)
g1_m;g2_m
# 표본의 크기
n1 <- length(data$group1)
n2 <- length(data$group2)
n1;n2
# 분산
var1 <- var(data$group1)
var2 <- var(data$group2)
var1;var2
# 등분산 가정
# 통합분산추정치 = (df1*var1 + df2*var2)/(df1+df2)
df1 <- n1 - 1
df2 <- n2 - 1
df1;df2
# 분자 = (df1*var1 + df2*var2)
(df1*var1 + df2*var2)
# 분모 = (df1+df2)
(df1+df2)
# 통합분산추정치 = 분자/분모 = sp2
sp2 <- (df1*var1 + df2*var2)/(df1+df2)
sp2
# 검정통계량 t값
# 분자 = g1_m - g2_m
(g1_m - g2_m)
# 분모 = 제곱근(sp2/n1 + sp2/n2)
sqrt(sp2/n1 + sp2/n2)
# 분자/분모
t <- (g1_m - g2_m)/sqrt(sp2/n1 + sp2/n2)
t <- abs(t)
t
# 임계값
# 유의수준 0.05, 양측검정, df1+df2 = n1 + n2 -2
alpha <- 0.05
df <- df1 + df2
alpha;df
cv <- qt(1 - alpha/2, df)
cv
cv < t
# FALSE = 채택역 = 귀무가설 채택 = 연구가설 기각 = 차이가 없다 = 통계적으로 유의한 차이가 없다
# 유의확률
p <- (1 - pt(t, df))*2
p
alpha > p
# FALSE = 채택역 = 귀무가설 채택 = 연구가설 기각 = 차이가 없다 = 통계적으로 유의한 차이가 없다
#
```

```python
# 짝지은 표본 t검정
pre <- c(3, 5, 4, 6, 5, 5, 4, 5, 3, 6, 7, 8, 7, 6, 7, 8, 8, 9, 9, 8, 7, 7, 6, 7, 8)
post <- c(7, 8, 6, 7, 8, 9, 6, 6, 7, 8, 8, 7, 9, 10, 9, 9, 8, 8, 4, 4, 5, 6, 9, 8, 12)
# 차이값, 평균, 표준편차, 표본의 크기
diff <- post - pre
diff
diff_m <- mean(diff)
diff_m
diff_sd <- sd(diff)
diff_sd
n <- length(diff)
n
# 검정통계량 t값
# 분자 = diff의 평균 - 0
diff_m
# 분모 = 표준오차 = diff의 표준편차/제곱근(n)
diff_sd/sqrt(n)
# 분자/분모
t <- diff_m/(diff_sd/sqrt(n))
t
# 임계값
# 유의수준, 양측검정, 자유도
alpha <- 0.05
df <- n - 1
cv <- qt(1 - alpha/2, df)
cv
cv < t
# TRUE = 오른쪽 = 기각역 = 귀무가설 기각 = 연구가설 채택 = 차이가 있다 = 통계적으로 유의한 차이가 있다
#
# 유의확률
pv <- (1 - pt(t, df))*2
alpha > pv
# TRUE = 오른쪽 = 기각역 = 귀무가설 기각 = 연구가설 채택 = 차이가 있다 = 통계적으로 유의한 차이가 있다
#
```

```python
# 단순 분산분석 = 일원배치 분산분석(one-way ANOVA)
x1 <- c(87, 86, 76, 56, 78, 98, 77, 66, 75, 67)
x2 <- c(87, 85, 99, 85, 79, 81, 82, 78, 85, 91)
x3 <- c(89, 91, 96, 87, 89, 90, 89, 96, 96, 93)
# 그룹 평균, 총 평균, 표본의 크기
x1_m <- mean(x1)
x2_m <- mean(x2)
x3_m <- mean(x3)
x1_m;x2_m;x3_m
total_m <- mean(c(x1,x2,x3))
total_m
n1 <- length(x1)
n2 <- length(x2)
n3 <- length(x3)
n1;n2;n3
# between: 그룹간 변동 = 그룹 평균 - 총 평균
b1 <- x1_m - total_m
b2 <- x2_m - total_m
b3 <- x3_m - total_m
b1;b2;b3
b1_sq <- b1**2
b2_sq <- b2**2
b3_sq <- b3**2
b1_sq;b2_sq;b3_sq
ssb <- sum(b1_sq*n1, b2_sq*n2, b3_sq*n3)
ssb
# within: 그룹 내 변동 = 관측치 - 평균
w1 <- x1 - x1_m
w2 <- x2 - x2_m
w3 <- x3 - x3_m
w1;w2;w3
w1_sq <- w1**2
w2_sq <- w2**2
w3_sq <- w3**2
w1_sq;w2_sq;w3_sq
ssw <- sum(w1_sq,w2_sq,w3_sq)
ssw
# between 자유도 = k - 1 (k: 그룹의 개수)
k = 3
b_df <- k - 1
b_df
# within 자유도 = n - k
n = n1 + n2 + n3
w_df = n - k
w_df
# MSB = SSB/(k - 1)
msb <- ssb/b_df
msb
# MSW = SSW/(n - k)
msw <- ssw/w_df
msw
# F 값 = MSB/MSW
f <- msb/msw
f
# 임계값, 유의수준, 자유도
alpha <- 0.05
b_df;w_df
cv <- qf(1 - alpha, b_df, w_df)
cv < f
# TRUE = 오른쪽 = 기각역 = 귀무가설 기각 = 연구가설 채택 = 차이가 있다 = 통계적으로 유의한 차이가 있다
#
# 유의확률
pv <- 1 - pf(f, b_df, w_df)
alpha > pv
# TRUE = 오른쪽 = 기각역 = 귀무가설 기각 = 연구가설 채택 = 차이가 있다 = 통계적으로 유의한 차이가 있다
#
```

```python
# 숫자 지수 표현 비활성화
options(scipen = 99)
# 상관계수
x = c(2, 4, 5, 6, 4, 7, 8, 5, 6, 7)
y = c(3, 2, 6, 5, 3, 6, 5, 4, 4, 5)
# x편차 = x값 - x평균
x_bar <- mean(x)
x_bar
x_dev <- x - x_bar
x_dev
# y편차 = y값 - y평균
y_bar <- mean(y)
y_bar
y_dev <- y - y_bar
y_dev
# 상관계수
# 분자 = x편차*y편차의 합
sum(x_dev*y_dev)
# 분모 = 제곱근(x편차제곱합*y편차제곱합)
# x편차제곱합
sum(x_dev**2)
# y편차제곱합
sum(y_dev**2)
# 분모
sqrt(sum(x_dev**2)*sum(y_dev**2))
# 분자/분모
sum(x_dev*y_dev)/(sqrt(sum(x_dev**2)*sum(y_dev**2)))
# r = 0.692 => 1) 양의 상관관계(직접상관관계), 2) 상관관계 크기가 강한 크기를 가지고 있음
#
# 산점도 = 상관관계의 시각화
plot(x, y)
#
```

```python
# 상관계수 t검정
x <- c(76, 81, 78, 76, 76, 78, 76, 78, 98, 88,
       76, 66, 44, 67, 65, 59, 87, 77, 79, 58,
       68, 76, 77, 98, 98, 99, 98, 87, 67, 78)
y <- c(43, 33, 23, 34, 31, 51, 56, 43, 44, 45,
       32, 33, 28, 39, 31, 38, 21, 27, 43, 46,
       41, 41, 48, 56, 56, 55, 45, 68, 54, 33)
# 산점도: 상관관계 시각화
plot(x, y)
# 상관계수
r <- cor(x,y)
r
# 검정통계량 t값 = (표본상관계수 - 모상관계수)/제곱근((1-표본상관계수의 제곱)/(표본의 크기-2))
# 분자 = (표본상관계수 - 모상관계수=0)
r
# 분모 = 제곱근((1-표본상관계수의 제곱)/(표본의 크기-2))
n <- length(x)
n
sqrt((1-r**2)/(n-2))
# 분자/분모
t <- r/sqrt((1-r**2)/(n-2))
t <- abs(t)
t
# 임계값, 자유도, 유의수준, 양측검정
df <- n - 2
alpha <- 0.05
cv <- qt(1-alpha/2, df);cv
cv < t
# TRUE = 오른쪽 = 기각역 = 귀무가설 기각 = 연구가설 채택 = 상관계수가 0이 아니다 = 통계적으로 유의하다
# 유의확률
p <- (1-pt(t, df))*2;p
alpha > p
# TRUE = 오른쪽 = 기각역 = 귀무가설 기각 = 연구가설 채택 = 상관계수가 0이 아니다 = 통계적으로 유의하다
#
# 결혼생활의 질과 부모관계의 질은 통계적으로 유의한 상관계수를 가지고 있다. 이 상관계수는 양의 상관관계(직접 상관관계)를 나타내고, 약한 크기의 상관관계를 가지고 있다고 해석할 수 있다.
#
```

```python
# 숫자 지수 표현 비활성화
options(scipen = 99)
# 선형회귀분석
# 단순선형회귀 simple regression
# 독립변수(x변수) 1개, 종속변수(y변수) 1개
x <- c(3.5, 2.5, 4, 3.8, 2.8, 1.9, 3.2, 3.7, 2.7, 3.3)
y <- c(3.3, 2.2, 3.5, 2.7, 3.5, 2, 3.1, 3.4, 1.9, 3.7)
# 평균
x_bar <- mean(x)
y_bar <- mean(y)
x_bar;y_bar
# 편차
# x편차 = x값 - x평균
x_dev <- x - x_bar
x_dev
# y편차 = y값 - y평균
y_dev <- y - y_bar
y_dev
# 기울기 = x편차*y편차 합/x편차제곱합
# 분자 = x편차*y편차 합
sum(x_dev*y_dev)
# 분모 = x편차제곱합
sum(x_dev**2)
# 분자/분모
b <- sum(x_dev*y_dev)/sum(x_dev**2)
b
# 절편 = y평균 - 기울기*x평균
a <- y_bar - b*x_bar
a
# 예측치 y hat
# x = 2
# y_hat <- b*x + a
# y_hat
# 회귀계수(=기울기)에 대한 t검정
# 검정통계량 t값 = 기울기/표준오차
# 분자 = 기울기
b
# 분모 = 표준오차 = 제곱근((y hat 편차제곱합/자유도)/x편차제곱합)
# y hat 편차 = y값 - y hat
# y hat = y 예측치 계산
y_hat <- b*x + a
y_hat
y_hat_dev <- y - y_hat
y_hat_dev
# y hat 편차제곱합
sum(y_hat_dev**2)
# 자유도 = n - k -1 (k: 독립변수의 개수)
k <- 1
n <- length(x)
df <- n - k - 1
k;n;df
# 분자 = (y hat 편차제곱합/자유도)
(sum(y_hat_dev**2)/df)
# 분모 = x편차제곱합
sum(x_dev**2)
# 제곱근(분자/분모)
se <- sqrt((sum(y_hat_dev**2)/df)/sum(x_dev**2))
se
# 검정통계량 t값 = 기울기/표준오차
t <- b/se
t <- abs(t)
t
# 임계값, 유의수준, 자유도, 양측검정
alpha <- 0.05
df
cv <- qt(1-alpha/2, df)
cv
cv < t
# TRUE = 오른쪽 = 기각역 = 귀무가설 기각 = 연구가설 채택 = 회귀계수가 0이 아니다 = 통계적으로 유의하다
# 유의확률
p <- (1 - pt(t, df))*2
alpha > p
# TRUE = 오른쪽 = 기각역 = 귀무가설 기각 = 연구가설 채택 = 회귀계수가 0이 아니다 = 통계적으로 유의하다
#
# 결정계수 = SSR/SST(=SSR+SSE)
# SSE
sse = sum(y_hat_dev**2)
sse
# SSR
# 회귀편차 = y hat - y평균
(y_hat - y_bar)
# 회귀편차제곱합 SSR
ssr <- sum((y_hat - y_bar)**2)
# 결정계수 = ssr/sst
ssr/(ssr+sse)
1-(sse/(ssr+sse))
# 46% 설명할 수 있다
# 모형 적합성 F검정
# 검정통계량 F값 = MSR/MSE
# MSR = SSR/k
msr <- ssr/k
msr
# MSE = SSE/(n-k-1)
mse <- sse/df
mse
# F값 = MSR/MSE
f <- msr/mse
f
# 임계값, 유의수준, 자유도(2개)
k;df
alpha <- 0.05
cv <- qf(1-alpha, k, df)
cv
cv < f
# TRUE = 오른쪽 = 기각역 = 귀무가설 기각 = 연구가설 채택 = 통계적으로 유의하다 = 모형이 적합하다
# 유의확률
p <- 1-pf(f, k, df)
p
alpha > p
# TRUE = 오른쪽 = 기각역 = 귀무가설 기각 = 연구가설 채택 = 통계적으로 유의하다 = 모형이 적합하다
#
sr <- lm(y~x)
summary(sr)
#
```

```python
# 독립성 검정
# row = 투표참여여부
# col = 성별
row1 <- c(37, 32)
row2 <- c(20, 31)
col1 <- c(37, 20)
col2 <- c(32, 31)
# 행합계
r1 <- sum(row1)
r2 <- sum(row2)
r1;r2
# 열합계
c1 <- sum(col1)
c2 <- sum(col2)
c1;c2
# 총 빈도
total <- sum(row1, row2)
total
# 기대빈도
e11 <- (r1*c1)/total
e12 <- (r1*c2)/total
e21 <- (r2*c1)/total
e22 <- (r2*c2)/total
e11;e12;e21;e22
# 검정통계량 카이제곱값
# (O - E)**2/E 의 합
chi <- sum((row1[1] - e11)**2/e11,
           (row1[2] - e12)**2/e12,
           (row2[1] - e21)**2/e21,
           (row2[2] - e22)**2/e22)
chi
# 임계값, 유의수준, 자유도 = (행개수-1)*(열개수-1)
alpha <- 0.05
df <- 1
df
cv <- qchisq(1-alpha, df)
cv
cv < chi
# FALSE = 채택역 = 귀무가설 채택 = 연구가설 기각 = 차이가 없다 = 통계적으로 유의한 차이가 없다 = 독립이다
# 유의확률
p <- 1-pchisq(chi, df)
p
alpha > p
# FALSE = 채택역 = 귀무가설 채택 = 연구가설 기각 = 차이가 없다 = 통계적으로 유의한 차이가 없다 = 독립이다
#
```
